```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("/content/hr_data.csv")
```

```python
data.head(3)
```

```python
data.info()
```

```python
data.isnull().sum()
```

To find females candidates have joined the orginsation beyond 30 notice period

```python
data[(data.Status == 'Joined') & (data.Gender == 'Female') & (data['Notice period'] > 30)]
```

To find males candidates above the age 35yrs are willing to recolate.

```python
data[(data.Gender=='Male')& (data.Age.astype(str)>'35') &(data['Candidate relocate actual']=="Yes") ]
```

To find if "Percent hike expected in CTC" and "Percent hike offered in CTC" affect joining

```python
sns.barplot(x="Status", y="Percent difference CTC" ,data = data ,hue="Gender")
```

To find what sources are used for recruitment of the candidates.

```python
sns.displot(x="Candidate Source", data= data)
```

Which Offered band has most no.of vancanies?

```python
data[ "Offered band"].max()
```

Average notice period for E2 band given joining bonus

```python
data.groupby("Notice period").mean()
```

Plottig a graph to find which Employee Band has most opening against each Department

```python
plt.figure(figsize=(20, 6))

plt.subplot(1, 2, 1)
sns.countplot(x='LOB', hue='Offered band', data= data, palette='Set2')
plt.title('Employee Band Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.legend(title='Band')

```

Plotting a graph to identify which deparment has better gender ratio.

```python

plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 1)

sns.countplot(x='LOB', hue='Gender', data=data)
plt.title('Gender Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.legend(title='Gender')

plt.tight_layout()
plt.show()
```
